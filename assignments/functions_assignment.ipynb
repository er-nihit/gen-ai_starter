> **Task 1: Basic Function (Price after discount)**


```python
# Function to apply discount b taking price and discount_percent as input.
# Default value of discount_percent is set to 5%
def apply_discount(price, discount_percent=5):
    if discount_percent>60: ## Additional check if the discount percent is more than 60, it will give max 60% discount
        discount_percent=60
    return price-price*discount_percent/100

print(apply_discount(1000,10))  # Example 1 - Normal usage
print(apply_discount(500))      # Example 2 - Without passing discount percent argument
print(apply_discount(10000,80)) # Example 3 - Passing discount_percent more than 60
```

> **Task 3: Lamba Function (GST Calculator)**

```python
# Defining lambda function to add GST
gst = lambda price:price+(0.18*price)

# Calling lambda function
print(gst(100))

# Creating new lambda function to add GST and discount together
discount = lambda price,discount:price+(0.18*price)-(discount*price*0.01)

# Calling new lambda function
print(discount(100,10))
```

> **Task 4: Using Map() - Apply GST to list of prices**

```python
# Using GST lamba from earlier cell to pass through map to iterate through list
prices = [100,250,400,1200,50]
print("Original Prices:",prices)
print("Price after GST",list(map(gst, prices)))
```

> **Task 5: Using filter() - Filter Expensive Products**

```python
# Lambda Function to filter check prices more than 500
expensive = lambda price:price>500

# Lamba function to filter check price less than or equal to 500
cheap = lambda price:price<=500


prices = [100,250,400,1200,50,2000,850]

# Using filter() to to filter out specific products
print("Expensive Prices:",list(filter(expensive,prices)))
print("Budget Friendly Prices:", list(filter(cheap,prices)))
```

> **Task 6: Combined Utility Function**

```python
# User defined function to do all calculations
def process_prices(prices):
    apply_discount = lambda price:price*0.9 # Lambda function to apply discounts
    discounted_price = list(map(apply_discount,prices)) # map discounts all prices
    filter_discounted = lambda final:final>300 #Lambda function to check if the discounted price is more than 300
    filtered_prices = list(filter(filter_discounted,discounted_price)) # Filter function to filter all discounted prices more than 300 using previous filter_discount lambda function
    return discounted_price,filtered_prices

prices = [100,500,900,50,750]
discounted_prices, filtered_prices = process_prices(prices)

print("All Discounted prices:",discounted_prices)
print("Dicounted prices more than 300:",filtered_prices)
```

> **Task 7: Mini Problem - Menu Using Functions**

```python
def add_price(prices_list, price):
    return prices_list.append(price)

def get_average_price(prices_list):
    return sum(prices_list)/len(prices_list)

def get_max_price(prices_list):
    return max(prices_list)

def loop(prices_list):
    user_input = input("1. Add Price\n2. Show Average Price\n3. Show Highest Price\nq. Quit")
    if user_input == '1':
        prices_list = add_price(prices_list,float(input("Enter value=")))
        print ("Updated List:",prices_list)
        return loop(prices_list)
    elif user_input == '2':
        print ("Average:", get_average_price(prices_list))
        return loop(prices_list)
    elif user_input == '3':
        print("Maximum Price:",get_max_price(prices_list))
        return loop(prices_list)
    elif user_input == 'q':
        print("Bye")
        return
    else:
        print("Incorrect input. Try again.")
        return loop(prices_list)
    
prices_list = [100,200,500]
loop(prices_list)
```

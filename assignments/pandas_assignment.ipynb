> **Task 1: Pandas Series Basics**

```python
import pandas as pd

marks = [78,85,90,66,72]
marks_series = pd.Series(marks)

print("Series values:",marks_series.values)
print("Series Index:",marks_series.index)
print("Data Types:",marks_series.dtypes)
print("First Element:",marks_series[0])
print("Last two elements:",marks_series.values[-2:])
```

```python
print("Add 5 grace marks:",marks_series + 5)
print("Reduce 2 marks:",marks_series - 2)
marks_float = pd.to_numeric(marks_series)
print("Multiply all marks by 1.05:",marks_float * 1.05)
print("Divide all marks by 2:", marks_float / 2)
```

```python
print("Maximum Marks:",marks_series.max())
print("Minimum Marks:",marks_series.min())
print("Sum of all marks:",marks_series.sum())
print("Mean of all marks:",marks_series.mean())

print("Student scored 70?",list(map(lambda x:x>=70,marks_series)))
print("Student who passed:",list(filter(lambda x:x>=70,marks_series)))
```

```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}

df = pd.DataFrame(students)
print("Dataframe:",df)
print("First 3 rows:",df.head(3))
print("Last 2 elements:",df.tail(2))
```

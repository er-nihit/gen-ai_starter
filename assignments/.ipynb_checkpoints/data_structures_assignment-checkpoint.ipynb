> **Task 1: Product Collections (Lists and Tuples)**

```python
# 1. Create a list named products containing at least 6 product names
products = ["Laptop", "Mobile Phone", "Headphones", "Keyboard", "Mouse", "Monitor"]
print("Product list:", products)

# 2. Create a tuple named sample_product with (product_name, price, category)
sample_product = ("HP Elitebook", 60000.00, "Laptop")
print("Sample product (tuple):", sample_product)

# 3. Print the 2nd and last product from the products list
print("2nd product from the list:", products[1])
print("Last product from the list:", products[-1])

# 4. Append two new product names to products and then print the updated list
products.append("Webcam")
products.append("USB Cable")
print("Updated products list after appending:", products)

# 5. Extra: Convert sample_product into a list, change its price, and convert back to tuple
# Convert tuple to list
sample_product_list = list(sample_product)
print("Confirming type: ",type(sample_product_list))
sample_product_list[1] = 70000.00

# Convert back to tuple
sample_product_tup = tuple(sample_product_list)
print("Converted back to tuple:", type(sample_product_tup))
```

> **Task 2: Categories (Sets)**

```python
# 1. Create a set of categories list of matching products
categories = ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Accessories", "Accessories"]
categories_set = set(categories)
print("Categories set:", categories_set)

# 2. Add a new category to the set
categories_set.add("Groceries")
print("Adding 'Groceries':", categories_set)
# Adding duplicate category
categories_set.add("Electronics")  
print("After adding duplicate category 'Electronics':", categories_set)

# 3. Check whether a category exists in the set (membership test returns boolean)
print("Checking exisitng category (Electronics): ", "Electronics" in categories_set)
print("Checking non-existing category (Clothing):", "Clothing" in categories_set)

# 4. Extra: Get the total number of unique categories and print them all
print("Total unique categories:", len(categories_set))
```

> **Task 3: Product Pricing (Dictionary)**

```python
# 1. New dictionary with product names as keys and prices as values
price_dict = {
    "Laptop": 60000.00,
    "Mobile Phone": 25000.00,
    "Headphones": 5000.00,
    "Keyboard": 3500.00,
    "Mouse": 1500.00,
    "Monitor": 18000.00
}
print(price_dict)

# 2. Editing dictionary
# Add a new product with price
price_dict["Webcam"] = 4500.00
print(price_dict)

# Update the price of an existing product
print("Old Laptop price: ", price_dict["Laptop"])
price_dict["Laptop"] = 65000.00
print("Updated Laptop price: ", price_dict["Laptop"])

# Removing a product using pop to avoid KeyError
removed_price = price_dict.pop("Charger", None)
if removed_price is None:
    print("'Charger' does not exist in dictionary")
else:
    print("Removed 'Charger' of Rs. ",removed_price)

# 3. Calculate average price
average_price = sum(price_dict.values()) / len(price_dict)
print("Average price of all products: Rs.",average_price)

# 4. Extra: Print product with maximum and minimum prices
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print("Product with maximum price:",max_product,"- Rs.",price_dict[max_product])
print("Product with minimum price:",min_product,"- Rs.",price_dict[min_product])
```

> **Task 4: Combined Operations**

```python
# 1. Creating a catalog with product_name, price, category
# Ensure we have products and their prices
catalog = []
for i, product in enumerate(products):
    if product in price_dict:
        price = price_dict[product]
        category = categories[i]
        catalog.append((product, price, category))

print("Catalog (product_name, price, category):")
for item in catalog:
    print(item)

# 2. Creating category_to_products dictionary mapping categories to product lists
category_to_products = {}
for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)

print("\nCategory to Products mapping:\n",category_to_products)

# 3. Finding the category with maximum number of products and print all products in it
max_category = max(category_to_products, key=lambda cat: len(category_to_products[cat]))
max_count = len(category_to_products[max_category])
products_in_max_category = category_to_products[max_category]

print("\nCategory with maximum products: ",max_category,"-",max_count)
print("Products in ",max_category,"category:")
for product in products_in_max_category:
    print(product,"-",price_dict[product])
```

> **Task 1: Pandas Series Basics**

```python
import pandas as pd

marks = [78,85,90,66,72]
marks_series = pd.Series(marks)

print("Series values:",marks_series.values)
print("Series Index:",marks_series.index)
print("Data Types:",marks_series.dtypes)
print("First Element:",marks_series[0])
print("Last two elements:",marks_series.values[-2:])
```

> **Task 2: Mathematical Operations on Series**

```python
print("Add 5 grace marks:",marks_series + 5)
print("Reduce 2 marks:",marks_series - 2)
marks_float = pd.to_numeric(marks_series)
print("Multiply all marks by 1.05:",marks_float * 1.05)
print("Divide all marks by 2:", marks_float / 2)
```

> **Task 3: Python Functionalities on Series**

```python
print("Maximum Marks:",marks_series.max())
print("Minimum Marks:",marks_series.min())
print("Sum of all marks:",marks_series.sum())
print("Mean of all marks:",marks_series.mean())

print("Student scored 70?",list(map(lambda x:x>=70,marks_series)))
print("Student who passed:",list(filter(lambda x:x>=70,marks_series)))
```

> **Task 4: Create a DataFrame**

```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}

df = pd.DataFrame(students)
print("Dataframe:",df)
print("First 3 rows:",df.head(3))
print("Last 2 elements:",df.tail(2))
print("Shape of Dataframe:",df.shape)
print("Columns in Dataframe:",df.columns)
```

> **Task 5: Important DataFrame Functions**

```python
print("INFO:\n",df.info())
print("\nDESCRIBE:\n",df.describe())
print("\nHEAD:\n",df.head(1))
print("\nTAIL:\n",df.tail(1))
sorted = df.sort_values(['Marks'], ascending=False)
print("\nSorted by Marks:\n",sorted)
print("\nIndex Reset:\n", sorted.reset_index().drop(['index'], axis=1))
```

> **Task 6: Filtering and Conditional Selections**

```python
print("Students who scored more than 75 marks:",', '.join(df[df['Marks']>75]['Name']))
print("Students who took maths:", ', '.join(df[df['Subject']=='Math']['Name']))
print("Students who scored more than average:",', '.join(df[df['Marks']>df['Marks'].mean()]['Name']))
print("Students who failed:",', '.join(df[df['Marks']<75]['Name']))
```

> **Task 7: Grouping and Basic Analysis**

```python
print("\nAverage Marks for each subject:\n",df.groupby('Subject', as_index=False)['Marks'].mean())
print("\nStudents in each subject:\n",df.groupby('Subject', as_index=False)['Name'].count())
print("\nMaximum marks in each subject:\n",df.groupby('Subject', as_index=False)['Marks'].max())
```

> **Task 8: Pandas Plotting (Simple graphs)**

```python
df.plot(x='Name', y='Marks', kind='bar', title='Bar graph')
df.plot(x='Name', y='Marks', kind='line', title='Line Graph')
df.plot(y='Marks', kind='hist', title='Histogram')
```

> **Task 9: Mini Use Case: Sales Data Analysis**

```python
sales = {
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}
sales_df = pd.DataFrame(sales)
print("Total revenue:",sales_df['Revenue'].sum())
print("Average daily revenue:",sales_df['Revenue'].mean())
print("Day with highest revenue:",sales_df.loc[sales_df['Revenue'].idxmax(), 'Day'])
print("Days with more than average sales:",', '.join(sales_df[sales_df['Revenue']>=sales_df['Revenue'].mean()]['Day']))
sales_df.plot(x='Day', y='Revenue', kind='bar', title='Bar Graph')
```

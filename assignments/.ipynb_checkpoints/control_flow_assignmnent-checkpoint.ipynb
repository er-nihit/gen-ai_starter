> **Task 1: Discount Rules**

```python
try:
    order_amount = int(input("Enter order amount: "))
except ValueError:
    print("Error: please enter a numeric order amount.")
    raise SystemExit

if order_amount >= 2000:
    discount_rate = 0.15
elif order_amount >= 1500:
    discount_rate = 0.10
elif order_amount >= 1000:
    discount_rate = 0.07
else:
    discount_rate = 0.0

discount = order_amount * discount_rate
subtotal = order_amount - discount
tax = subtotal * 0.05
final_total = subtotal + tax

print("Discount rate:", int(discount_rate * 100), "%")
print("Subtotal after discount:", subtotal)
print("Tax (5%):", tax)
print("Final total:", final_total)
```

> **Task 2: Process Multiple Orders**

```python
orders = [1200, 2500, 800, 1750, 3000]
total_revenue = 0
discounted_count = 0

print("order_amount -> discount% -> final_amount")

for order_amount in orders:
    if order_amount >= 2000:
        discount_rate = 0.15
    elif order_amount >= 1500:
        discount_rate = 0.10
    elif order_amount >= 1000:
        discount_rate = 0.07
    else:
        discount_rate = 0.0

    discount = order_amount * discount_rate
    final_amount = order_amount - discount
    total_revenue += final_amount

    if discount_rate > 0:
        discounted_count += 1

    print(order_amount, "->", int(discount_rate * 100), "%", "->", final_amount)

print("Total revenue after discounts:", total_revenue)
print("Orders with discount:", discounted_count)
```

> **Task 3: User Menu**

```python
orders = []

while True:
    print("\nMenu")
    print("1 - Add order amount")
    print("2 - Show all orders and totals after discounts")
    print("q - Quit")

    choice = input("Choose an option: ")

    if choice == "1":
        amount_text = input("Enter order amount: ")
        try:
            amount = int(amount_text)
        except ValueError:
            print("Invalid amount. Please enter a number.")
            continue

        orders.append(amount)
        print("Order added.")

    elif choice == "2":
        if len(orders) == 0:
            print("No orders yet.")
            continue

        total_revenue = 0
        print("order_amount -> discount% -> final_amount")

        for order_amount in orders:
            if order_amount >= 2000:
                discount_rate = 0.15
            elif order_amount >= 1500:
                discount_rate = 0.10
            elif order_amount >= 1000:
                discount_rate = 0.07
            else:
                discount_rate = 0.0

            discount = order_amount * discount_rate
            final_amount = order_amount - discount
            total_revenue += final_amount

            print(order_amount, "->", int(discount_rate * 100), "%", "->", final_amount)

        print("Total revenue after discounts:", total_revenue)

    elif choice == "q":
        print("Goodbye!")
        break

    else:
        print("Invalid option. Try again.")
        continue
```

> **Task 4: Loop Control with Conditions**

```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for amount in daily:
    if amount == -1:
        print("Corrupted data found. Stopping.")
        break
    if amount == 0:
        print("No sales day. Skipping.")
        continue

    total_sales += amount
    print("Running total:", total_sales)

print("Final total:", total_sales)
```

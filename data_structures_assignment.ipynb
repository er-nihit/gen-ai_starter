> **Task 1: Product Collections (Lists and Tuples)**

```python
# 1. Create a list named products containing at least 6 product names
products = ["Laptop", "Mobile Phone", "Headphones", "Keyboard", "Mouse", "Monitor"]
print("Product list:", products)

# 2. Create a tuple named sample_product with (product_name, price, category)
laptop_tup = ("HP Elitebook", 60000.00, "Laptop")
print("Sample product (tuple):", laptop_tup)

# 3. Print the 2nd and last product from the products list
print("2nd product from the list:", products[1])
print("Last product from the list:", products[-1])

# 4. Append two new product names to products and then print the updated list
products.append("Webcam")
products.append("USB Cable")
print("Updated products list after appending:", products)

# 5. Extra: Convert sample_product into a list, change its price, and convert back to tuple
# Convert tuple to list
laptop_list = list(laptop_tup)
print("Confirming type: ",type(laptop_list))
laptop_list[1] = 70000.00

# Convert back to tuple
laptop_tup = tuple(laptop_list)
print("Converted back to tuple:", type(laptop_tup))
```

> **Task 2: Categories (Sets)**

```python
# 1. Create a set of categories from a parallel list matching products
categories = ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Accessories", "Accessories"]
categories_set = set(categories)
print("1. Categories set created:", categories_set)

# 2. Add a new category to the set
categories_set.add("Home & Garden")
print("2. After adding 'Home & Garden':", categories_set)

# 2b. Demonstrate that duplicates are ignored
categories_set.add("Electronics")  # Try to add duplicate
print("   After trying to add duplicate 'Electronics' (ignored):", categories_set)

# 3. Check whether a category exists in the set (membership test returns boolean)
print("\n3. Membership checks:")
print("   Does 'Electronics' exist?", "Electronics" in categories_set)
print("   Does 'Clothing' exist?", "Clothing" in categories_set)

# Extra: Get the total number of unique categories and print them all
print("\n4. Extra - Total unique categories:", len(categories_set))
print("   All unique categories (sorted):", sorted(categories_set))
```
